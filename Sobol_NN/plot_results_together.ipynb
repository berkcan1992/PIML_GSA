```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
# from matplotlib.ticker import FormatStrFormatter
import os

# Adjust your matplotlib script by adding the following lines after import matplotlib
matplotlib.use("pdf")
# matplotlib.use("pgf")

    
matplotlib.rcParams.update({
        "pgf.texsystem": "pdflatex",
        'font.family': 'serif',
        'text.usetex': True,
        'pgf.rcfonts': False,
    })
# add LaTeX on python path
user_name = os.getlogin()
os.environ["PATH"] += os.pathsep + 'C:/Users/' + user_name + '/AppData/Local/Programs/MiKTeX 2.9/miktex/bin/x64'

#===========================     Using LaTeX compatible fonts      =============================== #
# use LaTeX fonts in the plot
plt.rc('text', usetex=True)
plt.rc('font', family='serif')    
```

```python

case = ["DNN","DNN_loss","DNN_hyb",
        "DNN_loss_hyb","DNN_upd_hyb","DNN_upd_loss",
        "DNN_upd_loss_hyb"]

# get the figure
fig = plt.figure()
    
for j,i in enumerate(case):
    results_dir = "../results/"
    filename = results_dir + str(i) + ".csv"
    data = pd.read_csv(filename, header=None, index_col=None)
    data = data.iloc[1:,:]
    data_np = data.to_numpy(dtype='float32')
    nobs = data_np[:,0]
    data_np = data_np[:,1:]
    data_np = np.around(data_np,6)

    sobol = "S1"

    colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

    plt.plot(nobs, data_np[:,0], '-o', color=colors[j])
    # plt.plot(nobs, data_np[:,2], '_', color=colors[1])
    # plt.plot(nobs, data_np[:,4], '_', color=colors[1])


    # plt.plot(obs[:,0], obs[:,1], 'o', color=colors[3])
    # plt.plot(pred_nn[:,0], pred_nn[:,1], '^', color=colors[0])
    # plt.plot(pred_nn_hyb[:,0], pred_nn_hyb[:,1], 'v', color=colors[1])
    # plt.plot(pred__piml[:,0], pred__piml[:,1], 'd', color=colors[4])
    # plt.plot(pred_piml_pre[:,0], pred_piml_pre[:,1], 's', color=colors[7])

plt.ylabel('$S_T$', fontsize=22)
plt.xlabel('n', fontsize=22)
plt.ylim((-0.01, 1))
plt.xlim((-0.01, 40))
# plt.xlim((-0.01, 1310))

plt.legend(['DNN', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}$', 'DNN$^{\mathrm{hyb}}$', 'DNN$^{\mathrm{hyb}, \mathcal{L}_{\mathrm{phy}}}$', 'DNN$^{\mathrm{upd, hyb}}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}$', 'DNN$^{\mathrm{upd, hyb}, \mathcal{L}_{\mathrm{phy}}}$'], loc='best', fontsize=12)
plt.tick_params(labelsize=20)
# ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.show()


plt_filename = '../figs/' +  sobol

# save as PDF
fig.savefig("{}.pdf".format(plt_filename), bbox_inches='tight', dpi=300)
```
